```python
#helper function to convert the string to list of dictionaries
import ast # converts string to list of dictionaries
def convert(obj):
    lst = []
    for i in ast.literal_eval(obj):
        lst.append(i['name'])
    return lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies
```

```python
def convert_key(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['keywords'] = movies['keywords'].apply(convert_key)
```

```python
movies
```

```python
# same thing for the cast and crew column
def convert_cast(obj):
    lst = []
    counter =0
    for i in ast.literal_eval(obj):
        if counter !=3:
            lst.append(i['name'])
            counter+=1
        else:
            break
    return lst
    
```

```python
# cast
movies['cast']  = movies['cast'].apply(convert_cast)
```

```python
#crew
# I want only director name from the crew column only
def convert_crew(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list
```

```python
#crew
movies['crew']   = movies['crew'].apply(convert_crew)
```

```python
# we have to convert the overview column into the list of words
movies['overview'] = movies['overview'].fillna('').apply(lambda x: x.split())
```

- Now we are going to perform the vectorisation on the tags columns of the movies dataset which further helps us in the finding the movie based on the similarity it uses the baging of words technique to transform the numerical data into the categorical data and by the help of these transformation it converts the data into the techniques perform the  steps like:
- first it remove the spaces from all the text
- then it convert the text into the lower format
- then it removes the stopping words which we also known as the helping verbs from the text like it,they etc.
- then it asssigns the token(1) to the words which are frequent in the text
- and token(0)which occurs one in the text.

```python
# for performing the text vectorization we use the sklearn library count vectorizer

from sklearn.feature_extraction.text import CountVectorizer
cv  = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
def stem(text):
    list = []
    for i in text.split():
        ps.stem(i)
    return " ".join(list)
```

Now all the movies is convertes into the vectors now it's time to use these vactors for finding the most related movies by using the cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0] # it is calculating the distance of all the movies that are realted to the first movie
# it is calacualting the distance between all the vectors after vectorization it converts all the words into vector by using the cosine similarity
```

Now I want to define a function which find's the given movie with the shortest value of the vector and return the movie name that is present on that vector

```python
def recommend(movie):
    movie_index = movies[movies['title']==movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
